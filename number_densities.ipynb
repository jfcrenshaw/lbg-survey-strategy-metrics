This notebook plots the luminosity functions

```python
import numpy as np
import matplotlib.pyplot as plt
from lbg_survey_metrics.constants import *
from lbg_survey_metrics.galaxy_distribution import *
```

```python
fig, ax = plt.subplots(figsize=single_col, constrained_layout=True, dpi=150)

M = np.linspace(-24, -17, 1000)

for band, z in zip("ugriz", np.arange(3, 8)):
    lf_u = dpl_lf(M, z)
    ax.plot(M, lf_u, c=colors[band], label=f"${band}$ ($z \sim {z}$)")


ax.legend()
ax.set(
    yscale="log",
    xlim=(M.min(), M.max()),
    ylim=(1e-8, 1e-2),
    xlabel="Absolute magnitude in detection band",
    ylabel="$\Phi$ (mag$^{-1}$ Mpc$^{-3})$",
    #xticks=np.arange(22, 30, 2),
    #yticks=[1e-10, 1e-8, 1e-6, 1e-4, 1e-2],
)

fig.savefig("figures/luminosity_functions.pdf")
```

```python
fig, ax = plt.subplots(figsize=single_col, constrained_layout=True, dpi=150)

m = np.linspace(23, 28, 100)

for band, z in zip("ugriz", np.arange(3, 8)):
    ax.plot(m, number_density(m, band), c=colors[band], label=f"${band}$ ($z \sim {z}$)")

ax.legend(title="Dropouts", handlelength=1.5)
ax.set(
    yscale="log",
    xlim=(m.min(), m.max()),
    ylim=(1e-3, 1e5),
    xlabel="Magnitude cut in detection band",
    ylabel="Number density (deg$^{-2}$)"
)

ax.scatter(23.79, number_density(23.79, "u"), c=colors["u"], marker="|", s=100)
ax.scatter(24.94, number_density(24.94, "u"), c=colors["u"], marker="o", s=15)
ax.scatter(26.8, number_density(26.8, "u"), c=colors["u"], marker="*", s=40)

ax.scatter(25.02, number_density(25.02, "g"), c=colors["g"], marker="|", s=100)
ax.scatter(26.17, number_density(26.17, "g"), c=colors["g"], marker="o", s=15)
ax.scatter(27.9, number_density(27.9, "g"), c=colors["g"], marker="*", s=40)

ax.scatter(24.55, number_density(24.55, "r"), c=colors["r"], marker="|", s=100)
ax.scatter(25.72, number_density(25.72, "r"), c=colors["r"], marker="o", s=15)
ax.scatter(27.4, number_density(27.4, "r"), c=colors["r"], marker="*", s=40)

ax.scatter(23.67, number_density(23.67, "i"), c=colors["i"], marker="|", s=100, zorder=10)
ax.scatter(24.78, number_density(24.78, "i"), c=colors["i"], marker="o", s=15, zorder=10)
ax.scatter(26.05, number_density(26.05, "i"), c=colors["i"], marker="*", s=40, zorder=10)

ax.scatter(23.67, number_density(23.67, "z"), c=colors["z"], marker="|", s=100, zorder=10)
ax.scatter(24.78, number_density(24.78, "z"), c=colors["z"], marker="o", s=15, zorder=10)
ax.scatter(26.05, number_density(26.05, "z"), c=colors["z"], marker="*", s=40, zorder=10)

fig.savefig("figures/number_densities.pdf")
```
