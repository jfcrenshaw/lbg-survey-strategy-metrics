This notebook plots the luminosity functions

```python
import numpy as np
import matplotlib.pyplot as plt
from src.constants import *
from src.number_density import *
```

```python
fig, ax = plt.subplots(figsize=single_col, constrained_layout=True, dpi=150)

m = np.linspace(23, 28, 1000)

lf_u = schecter_lf(m, **lf_params["u"], redshift=3)
ax.plot(m, lf_u, c=color_u, label="$u$ dropouts ($z \sim 3$)")

lf_g = schecter_lf(m, **lf_params["g"], redshift=4)
ax.plot(m, lf_g, c=color_g, label="$g$ dropouts ($z \sim 4$)")

lf_r = schecter_lf(m, **lf_params["r"], redshift=5)
ax.plot(m, lf_r, c=color_r, label="$r$ dropouts ($z \sim 5$)")

ax.legend()
ax.set(
    yscale="log",
    xlim=(m.min(), m.max()),
    ylim=(1e-8, 1e-2),
    xlabel="Apparent magnitude",
    ylabel="$\Phi$ (mag$^{-1}$ Mpc$^{-3})$",
    #xticks=np.arange(22, 30, 2),
    #yticks=[1e-10, 1e-8, 1e-6, 1e-4, 1e-2],
)

fig.savefig("figures/luminosity_functions.pdf")
```

```python
fig, ax = plt.subplots(figsize=single_col, constrained_layout=True, dpi=150)

ax.plot(m, number_density(m, "u", 3), c=color_u, label="$u$ dropouts ($z \sim 3$)")
ax.plot(m, number_density(m, "g", 4), c=color_g, label="$g$ dropouts ($z \sim 4$)")
ax.plot(m, number_density(m, "r", 5), c=color_r, label="$r$ dropouts ($z \sim 5$)")

ax.legend()
ax.set(
    yscale="log",
    xlim=(m.min(), m.max()),
    xlabel="5-sigma depth in detection band",
    ylabel="Number density (deg$^{-2}$)"
)

fig.savefig("figures/number_densities.pdf")
```
